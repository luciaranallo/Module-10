```python
# Based on the Attribute information I renamed each of the columns 
# I also added an entry column so that each row had a unique identifier to be used later

data.columns=['Class','Age','Menopause','Tumor Size','Inv-Nodes','Node-Caps','Deg-Malig','Breast','Breast-Quad','Irradiat']
data["Entry"] = range(0,len(data))
```

```python
data.head()
```

### Melt

```python
# melt the dataframe down to use the age column as the identifier and the corresponding values in the Menopause column
# this could be used to see the trends between age and when an individual enters menopause

data_melt = pd.melt(data, id_vars = ['Age'], value_vars = ['Menopause'])
```

```python
data_melt.head()
```

### Pivot

```python
# I pivoted the table to have the entry # as the row and the Breast as the 2 columns, with the tumor size as the value
# This table could be used to calculate aggregates on the different Breasts

data_pivot = data.pivot(index = 'Entry', columns = 'Breast', values = 'Tumor Size')
```

```python
data_pivot.head()
```

### Aggregation

```python
# Because most of the data is categorical there weren't many options to aggregate it.
# I chose to select the minimum and maximum of both the Age column and Tumor Size to show what the ranges are

data_aggregate = data.agg({'Age': ['min','max'], 'Tumor Size' : ['min','max']})
```

```python
data_aggregate
```

### Iteration

```python
# In this iteration I went through and printed each of the column names in the dataframe

for col in data:
    print(col)
```

```python
# I created a second smaller dataset that only includes the Entry #, the Age, and the Tumor size for the first 5 entries
# I then iterated through the table printing each row of the dataframe as a tuple

data2 = pd.DataFrame({"Entry" : data["Entry"][0:5], "Age" : data['Age'][0:5], "Tumor Size" : data['Tumor Size'][0:5]})

for row in data2.itertuples():
    print(row)
```

### Group By

```python
# I chose to group by both the Age and Menopause and then calculate the average within each group
# This calculates the average of each of the quantitative variables, and in this case Degree of Malignancy is the only one
# This table could be used to compare the Degrees of Malignancy both across age groups as well as menopause within

data_groupBy = data.groupby(['Age','Menopause']).mean()
```
